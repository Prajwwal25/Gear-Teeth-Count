```python


# Import necessary libraries
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os
from google.colab import files



# Define paths
model_path = '/content/datasets/Test-3/runs/detect/train2/weights/best.pt'  # Path to your trained model
input_image_path = '/content/datasets/Test-3/test/images/IMG_1286_JPG_jpg.rf.366a12c54842fa339a0c14041b29c410.jpg'  # Path to your input image

# Upload the input image to Colab
uploaded = files.upload()
for filename in uploaded.keys():
    input_image_path = f'/content/{filename}'

# Load the trained YOLOv11 model
model = YOLO(model_path)

# Perform inference on the input image
results = model.predict(source=input_image_path, conf=0.5)  # conf=0.5 is the confidence threshold

# Get the first result (assuming single image input)
result = results[0]

# Draw bounding boxes on the image
annotated_image = result.plot()  # This draws boxes on the image

# Convert BGR (OpenCV format) to RGB (Matplotlib format)
annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

# Display the image with bounding boxes
plt.figure(figsize=(10, 10))
plt.imshow(annotated_image)
plt.axis('off')
plt.show()

# Count and print the number of detected gear teeth
num_detections = len(result.boxes)
print(f"Number of detected gear teeth: {num_detections}")

# Optional: Save the annotated image
output_path = '/content/annotated_image.jpg'
cv2.imwrite(output_path, cv2.cvtColor(annotated_image, cv2.COLOR_RGB2BGR))
print(f"Annotated image saved to {output_path}")
```
